# 1 - Cleaning the .csv

```python
import pandas as pd

import sys
sys.path.append("../")

import re
import numpy as np
```

```python
royale = pd.read_csv("data/cardsInfo.csv",encoding = "ISO-8859-1")
```

```python
test = royale.apply(lambda x: "Common" if x.hitpoints1 else "Rare" if x.hitpoints3 else "Epic" if x.hitpoints6 else "Legendary" if x.hitpoints9 else "Spell", axis = 1)
```

```python
royale["damage14"]
```

```python
royale["max_damage_per_elixir"] = royale.apply(lambda x: round((x.damage14/x.elixir), 1) if x.elixir != 0 else 0, axis=1)
```

```python
royale["max_hitpoints_per_elixir"] = royale.apply(lambda x: round((x.hitpoints14/x.elixir), 1) if x.elixir != 0 else 0, axis=1)
```

```python
royale
```

```python
final_royale = royale[["name", "elixir", "cardtype", "hitpoints14", "max_hitpoints_per_elixir", "damage14", "max_damage_per_elixir"]]
```

```python
final_royale
```

```python
final_royale.to_csv("royale_clean", sep=',')
```

# 2 - Web Scraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
url = "https://statsroyale.com/decks/popular?type=tournament"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content)
```

```python
card_sets = soup.select("div.popularDecks_deckWrapper")
```

```python
len(card_sets)
```

```python
prueba=card_sets[0]
```

```python
lista=prueba.select("div.popularDecks__decklist")[0]
```

```python
t=lista.find_all("a")
```

```python
links=[]
for item in t:
    link = item.get("href")
    links.append(link)
```

```python
links
```

```python
deck = []
for link in links:
    separated = link.split("/")
    card = separated[-1]
    card = card.replace("+", " ")
    deck.append(card)
```

```python
deck
```

```python
wins = soup.select("div.popularDecks__footer div.ui__headerBig")[0]
float(wins.text.strip()[:-1])
```

```python
crowns = soup.select("div.popularDecks__footer div.ui__mediumText")[1]
ratiocrowns = float(crowns.text.split(" ")[0])
```

```python
def get_stats (soup):
    wins = soup.select("div.popularDecks__footer div.ui__headerBig")[0]
    ratiowin = float(wins.text.strip()[:-1])
    crowns = soup.select("div.popularDecks__footer div.ui__mediumText")[1]
    ratiocrowns = float(crowns.text.split(" ")[0])
    return [ratiowin, ratiocrowns]
```

```python
par = [deck, get_stats(card_sets[0])]
par
```

```python
df = pd.DataFrame(par)
```
